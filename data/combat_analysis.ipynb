```python
import pandas as pd
```

```python
data = pd.read_csv('combats.csv')
```

```python
pd.DataFrame.describe(data)
```

```python
first = data['First_pokemon']
second = data['Second_pokemon']
winner = data['Winner']
```

```python
def CountFrequency(my_list): 
  
    # Creating an empty dictionary  
    freq = {} 
    for item in my_list: 
        if (item in freq): 
            freq[item] += 1
        else: 
            freq[item] = 1
  
    for key, value in freq.items(): 
        print ("% d : % d"%(key, value)) 
  
# Driver function 
if __name__ == "__main__":  
    my_list = first
  
    CountFrequency(my_list) 
```
